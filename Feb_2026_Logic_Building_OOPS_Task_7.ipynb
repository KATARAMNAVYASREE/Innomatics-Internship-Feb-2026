# **Problem 1: Smart Light Controller**
## **Problem Statement**
## Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.is_on = False  # Default state is OFF

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        if self.is_on:
            print(f"{self.light_name} is ON")
        else:
            print(f"{self.light_name} is OFF")


# Taking input
light_name = input()
action = input()

# Creating object
light = SmartLight(light_name)

# Performing action
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

# Display result
light.display_status()
```

# **Problem 2: Employee ID Card System**
## **Problem Statement**
## Create a class that stores employee name, ID, and department and displays ID card details.


```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Taking formatted input with prompt shown
name_input = input("Employee Name: ")
id_input = input("Employee ID: ")
dept_input = input("Department: ")

# Direct values (no need to split now)
employee = Employee(name_input, id_input, dept_input)

employee.display_id_card()
```

# **Problem 3: Mobile Contact Record**
## **Problem Statement**
## Create a class to store contact name and phone number and display contact information.


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

# **Problem 4: Product Price Tag Generator**
## **Problem Statement**
## Create a class that stores product name and price and prints a formatted price tag.


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


# Input with labels shown while entering
name = input("Product Name: ")
price = input("Price: ")

# Creating object
product = Product(name, price)

# Displaying output
product.display_price_tag()
```

# **Problem 6: Delivery Address Manager**
## **Problem Statement**
## Create a class that stores customer name and delivery address and prints delivery details.


```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


name_input = input()
address_input = input()

customer_name = name_input.split(":")[-1].strip()
address = address_input.split(":")[-1].strip()

delivery = Delivery(customer_name, address)
delivery.display_details()
```
