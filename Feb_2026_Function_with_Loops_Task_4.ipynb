# **Problem 1: Mobile Recharge Validation System**
## **Problem Statement**
## Create a function to validate a mobile recharge request.

## **Rules:**
* Recharge amount must be ≥ ₹50
* Valid plans: 199, 299, 399, 599
* User should be allowed to retry if invalid








```python
# valid recharge plans
plans = [199, 299, 399, 599]

# function to check recharge
def check_recharge(amount):
  if amount < 50:
    print("Amount must be at least ₹50. Try again.\n")
    return False
  elif amount not in plans:
    print("Invalid plan! Available plans: 199, 299, 399, 599\n")
    return False

  return True

# retry using while loop
while True:
  amount = int(input("Enter recharge amount: ₹"))

  if check_recharge(amount):
    print("Recharge successful!")
    break
```

# **Problem 2: Inventory Reorder Alert System**
## **Problem Statement**
# Given a dictionary of products and stock quantities:
### * Print Reorder Alert if stock < 15
### * Print Stock OK otherwise








```python
# dictionary of products and stock
inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Salt": 8,
    "Oil": 15
}

# function to check stock
def check_stock(products):
  for item in products:
    if products[item] < 15:
      print(item, "-> Reorder Alert")
    else:
      print(item, "-> Stock OK")

# calling function
check_stock(inventory)
```

# **Problem 3: Student Result Processing System**
## **Problem Statement**
## Create a function that:
* Accepts a list of student marks
* Calculates average marks
* Prints Pass if average ≥ 50 else Fail





```python
# function to calculate result
def check_result(marks):

    total = 0

    # loop to calculate total marks
    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass", avg
    else:
        return "Fail", avg


# taking input from user
marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    m = int(input(f"Enter marks for subject {i+1}: "))
    marks.append(m)

# calling function
result, average = check_result(marks)

print("Average Marks:", average)
print("Result:", result)
```

#**Problem 4: Cab Fare Estimator with Retry Option**
## **Problem Statement**
## Create a cab fare calculator:
* Base fare = ₹50
* ₹12 per km
* Peak hour → 25% extra
* Allow retry using while loop

```python
# function to calculate fare
def calculate_fare(distance, peak):
  fare = 50 + (distance * 12)

  if peak == "yes":
    fare = fare + (fare * 0.25)

  return fare


# retry loop
while True:

  distance = float(input("Enter distance in km: "))

  if distance <= 0:
    print("Invalid distance! Try again.\n")
    continue

  peak = input("Is it peak hour? (yes/no): ").lower()

  if peak not in ["yes", "no"]:
    print("Please enter yes or no.\n")
    continue

  total_fare = calculate_fare(distance, peak)

  print("Total Fare: ₹", total_fare)
  break
```

# **Problem 5: Employee Attendance Eligibility Checker**
## **Problem Statement**
## Create a function that:
* Accepts attendance list ("P", "A")
* Calculates attendance percentage
* Returns Eligible if ≥ 75% else Not Eligible









```python
# function to check eligibility
def check_eligibility(attendance):
  present_days = 0

    # count present days
  for day in attendance:
    if day == "P":
      present_days += 1

  percentage = (present_days / len(attendance)) * 100

  if percentage >= 75:
    return "Eligible", percentage
  else:
    return "Not Eligible", percentage


# taking input
attendance = []
n = int(input("Enter number of working days: "))

for i in range(n):
  status = input(f"Day {i+1} (P/A): ").upper()
  attendance.append(status)

# function call
result, percent = check_eligibility(attendance)

print("Attendance Percentage:", percent)
print("Status:", result)
```

# **Problem 6: Password Strength Checker**
## **Problem Statement**
## Create a function to check password strength.
## **Rules:**
* Minimum length = 8
* Must contain at least one digit
* Must contain at least one special character (@#$)












```python
# function to check password strength
def check_password(password):
  has_digit = False
  has_special = False
  special_chars = "@#$"

  # check each character
  for ch in password:
    if ch.isdigit():
      has_digit = True
    if ch in special_chars:
      has_special = True

  # conditions
  if len(password) < 8:
    return "Weak Password (minimum 8 characters required)"

  if not has_digit:
    return "Weak Password (must include a number)"

  if not has_special:
    return "Weak Password (must include @, # or $)"

  return "Strong Password"


# input
pwd = input("Enter password: ")

# function call
result = check_password(pwd)

print(result)
```
