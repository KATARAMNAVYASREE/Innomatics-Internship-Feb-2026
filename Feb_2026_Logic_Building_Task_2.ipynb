## **Problem 1: Unique Words in a Sentence**
### **Problem Statement**
### Given a sentence, identify all unique words and display:
### 1. The count of unique words
### 2. The unique words themselves


```python
# Problem 1: Unique Words in a Sentence

import string

# Input and normalization
sentence = input("Enter sentence: ").lower()
sentence = sentence.translate(str.maketrans('', '', string.punctuation))  # remove punctuation

words = sentence.split()

unique_words = []
seen = set()   # used for fast duplicate checking

for word in words:
    if word not in seen:   # add only first occurrence
        unique_words.append(word)
        seen.add(word)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## **Problem 2: Highest Salary from Employee Data**
### **Problem Statement**
### Store employee names and their salaries, and determine which employee has the highest salary.


```python
# Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

## **Problem 3: Find Maximum and Minimum Values**
### **Problem Statement**
### Find the maximum and minimum values from a list of numbers.


```python
# Problem 3: Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

## **Problem 4: Count Products Above a Price Threshold**
### **Problem Statement**
### Count how many products have a price greater than 1000.


```python
# Problem 4: Products above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

## **Problem 5: Calculate Attendance Percentage**
### **Problem Statement**
### Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
# Problem 5: Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_days = attendance.count("P")
total_days = len(attendance)

percentage = (present_days / total_days) * 100

print("Attendance Percentage:", percentage)
```

## **Problem 7: Count Character Frequency**
### **Problem Statement**
### Count how many times each character appears in a given string.


```python
# Problem 7: Character Frequency

text = input("Enter a string: ")

freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

print(freq)
```
